# Checkpoint 5: Sector Analysis 2

```python
#C1 = 'USA' for FT = 'venture'
D1 = master_frame.loc[master_frame['country_code'].isin(['USA']) & master_frame['funding_round_type'].isin(['venture']), :]
#C2 = 'GBR' for FT = 'venture'
D2 = master_frame.loc[master_frame['country_code'].isin(['GBR']) & master_frame['funding_round_type'].isin(['venture']), :]
#C3 = 'IND' for FT = 'venture'
D3 = master_frame.loc[master_frame['country_code'].isin(['IND']) & master_frame['funding_round_type'].isin(['venture']), :]
```

```python
#The total number of investments for each main sector in a separate column
D1 = D1.assign(count_of_investments_usa = D1.groupby('main_sector')['raised_amount_usd'].transform('count'))

#The total amount invested in each main sector in a separate column
D1 = D1.assign(amount_of_investments_usa = round(D1.groupby('main_sector')['raised_amount_usd'].transform('sum')/10**6,2))
```

```python
D1.head()
```

```python
#The total number of investments for each main sector in a separate column
D2 = D2.assign(count_of_investments_gbr = D2.groupby('main_sector')['raised_amount_usd'].transform('count'))

#The total amount invested in each main sector in a separate column
D2 = D2.assign(amount_of_investments_gbr = round(D2.groupby('main_sector')['raised_amount_usd'].transform('sum')/10**6,2))
```

```python
#The total number of investments for each main sector in a separate column
D3 = D3.assign(count_of_investments_ind = D3.groupby('main_sector')['raised_amount_usd'].transform('count'))

#The total amount invested in each main sector in a separate column
D3 = D3.assign(amount_of_investments_ind = round(D3.groupby('main_sector')['raised_amount_usd'].transform('sum')/10**6,2))
```

```python
D = D1.append(D2.append(D3))
```
